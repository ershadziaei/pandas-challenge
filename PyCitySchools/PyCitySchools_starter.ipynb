# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending 645-675 per student actually underperformed compared to schools with smaller budgets (585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
#Get the datatypes
school_data_complete.head()
school_data_complete.dtypes
#take school data df and get the total number of schools
sum_school = school_data['school_name'].count()
#take student data df and count the number of students in total
sum_students = student_data['student_name'].count()
sum_students_final = f"{sum_students:,}"
#take the school data df and get the total budget
total_budget = school_data['budget'].sum()
#format budget figure
total_budget = (f'${total_budget:,}')
#take student data and get the average math score and round
avg_math = round(student_data['math_score'].mean(),6)
#take student data df and get the average reading score
avg_read = round(student_data['reading_score'].mean(),6)
#filter dataset to students who got 70 or higher
pass_math = (student_data.loc[(student_data['math_score']>=70)])
#take the number of students who passed math and divide by total number of students
pass_math_70 = round((len(pass_math)/sum_students)*100,6)
#filter students who got 70 or higher in reading
pass_reading = (student_data.loc[student_data['reading_score']>=70])
#take the total number of filtered students passing reading and divide by total number of students
pass_reading_70 = round((len(pass_reading) / sum_students)*100,6)
#take the students who passed math and reading with 70 or higher and filter them out
overall_pass = student_data.loc[(student_data['math_score']>=70) & (student_data['reading_score']>=70)]
#take overall pass filtered list, get number of students who passed both and divide by number of students
overall_pass = round((len(overall_pass)/sum_students)*100, 6)

#create new dataframe and populate with info above

student_summary_students = pd.DataFrame({'Total Schools': sum_school,
                                        'Total Students': sum_students_final,
                                        'Total Budget': total_budget,
                                        'Average Math Score': avg_math,
                                        'Average Reading Score': avg_read,
                                        '% Passing Math': pass_math_70,
                                        '% Passing Reading': pass_reading_70,
                                        '% Overall Passing': overall_pass},index=[0])
student_summary_students
```

## School Summary

```python
#school name
school_name = school_data.set_index(['school_name'])['type']
school_name
```

```python
#total number of students
total_students = school_data_complete['school_name'].value_counts()
total_students
```

```python
#take school data, set index to school name, and extract budget data by the school
budget_schools = school_data.set_index(['school_name'])['budget']#.map("${:,.2f}".format)
budget_schools
```

```python
#takes budget by the school and divides it by total number of studets
#getting buget by school per student
per_student_budget = budget_schools/total_students
per_student_budget
```

```python
#group school data by school name and run mean function on column math_score
avg_math_score = school_data_complete.groupby(['school_name'])['math_score'].mean()
avg_math_score
```

```python
#take merge data, group by school name, run mean on reading score
avg_reading_score = school_data_complete.groupby(['school_name'])['reading_score'].mean()
avg_reading_score
```

```python
#take merge data and show only students with math score 
math_pass = school_data_complete[(school_data_complete['math_score']>=70)]
math_pass.head()
```

```python
#take the df of students who passed path, group by school name, and count
school_math_pass = math_pass.groupby(['school_name'])['student_name'].count()
school_math_pass
```

```python
#create new dataframe from merged dataframe, filter only students who passed reading with 70 or higher
reading_pass = school_data_complete[(school_data_complete['reading_score']>=70)]
reading_pass
```

```python
#take the passed reading dataframe, groupby school name, and count the amount of students who passed
school_reading_pass = reading_pass.groupby(['school_name'])['reading_score'].count()
school_reading_pass
```

```python
pct_school_reading_pass = (school_reading_pass/total_students)*100
pct_school_reading_pass.sort_values(ascending=False)
```

```python
pct_school_math_pass = (school_math_pass/total_students)*100
pct_school_math_pass.sort_values(ascending=False)
```

```python
overall_pass = school_data_complete[(school_data_complete['reading_score']>=70) & (school_data_complete['math_score']>=70)]
overall_pass
```

```python
overall_pass_count = overall_pass.groupby(['school_name'])['student_name'].count()
overall_pass_count
```

```python
pct_overall_pass = (overall_pass_count/total_students)*100
pct_overall_pass.sort_values(ascending=False)
```

```python
new_df = pd.DataFrame({'School Type': school_name, 'Total Students': total_students, 'Total School Budget': budget_schools, 'Per Student Budget': per_student_budget,
                      'Average Math Score': avg_math_score, 'Average Reading Score': avg_reading_score,
                      '% Passing Math': pct_school_math_pass, '% Passing Reading': pct_school_reading_pass, '% Overall Passing': pct_overall_pass})
new_df
```
